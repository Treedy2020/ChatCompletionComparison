```python
import pandas as pd

result_4o = pd.read_csv('./data/result_gpt4o.csv')
result_4turbo = pd.read_csv('./data/result_gpt4turbo.csv')
```

```python
name_set = set(result_4o['fullname'])
result_4turbo = result_4turbo[result_4turbo['fullname'].isin(name_set)]
```

```python
result_4o.drop_duplicates(subset=['fullname'], inplace=True)
result_4turbo.drop_duplicates(subset=['fullname'], inplace=True)
```

```python
final_result = pd.merge(result_4o, result_4turbo, on=['fullname', 'full_address'])
```

```python
final_result.columns = ['fullname', 'full_address', '4o-response', '4turbo-response']

final_result.to_csv('./data/final_result.csv', index=False)
```

```python
print("gpt-4o 答案获取率:", round(1 - final_result['4o-response'].to_list().count("Unknown") / final_result['fullname'].count(), 4))
print("gpt-4turbo 答案获取率:", round(1 - final_result['4turbo-response'].to_list().count("Unknown") / final_result['fullname'].count(), 4))
```
